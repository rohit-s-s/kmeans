```python
n_a = 5
n_b = 4
n_c = 6
n_d = 4
n_e = 4
n_g = 4
n_f = 6
n_h = 4
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# Specify the number of clusters and create a KMeans model
num_clusters = n_a
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(a)

# Get cluster labels and cluster centers
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

# Visualize the clusters
for i in range(num_clusters):
    plt.scatter(a[cluster_labels == i][:, 1], a[cluster_labels == i][:, 0], label=f'Cluster {i + 1}')

plt.scatter(cluster_centers[:, 1], cluster_centers[:, 0], c='black', marker='x', s=100, label='Cluster Centers')  # Corrected line
plt.xlabel('Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('K-Means Clustering of locations')
plt.legend()
plt.show()
n_of_a = cluster_centers
```
